```python
import sys
sys.path.insert(0, '/Users/lelouch/pyw')
from pyw.utils.predicates import *

from pyw.core import AffineLieAlgebra, get_marks, get_comarks
```

```python
# A2^(1) affine Lie algebra - simply laced
alg = AffineLieAlgebra(['A', 2, 1])
```

```python
tt = -9
u = 4
delta = alg.delta()
rho_hat = alg.rho_hat()
theta_hat = alg.affine_theta()
Pi_u = [u * delta - theta_hat] + [alg.affine_simple_roots()[i] for i in range(1, alg.rank)]

W_elements = alg.extended_affine_weyl_group().elements_as_semi_direct_product(
    translation_bounds={1: (-10, 11), 2: (-10, 11)})

# 筛选出所有元素都是 True 的项，并找出对应的 w
results = [[is_positive_affine_root(w.action(weight))
            for weight in Pi_u] for w in W_elements]
valid_pairs = [(w, result)
               for w, result in zip(W_elements, results) if all(result)]

print(f"找到 {len(valid_pairs)} 个满足条件的 w:")
for w, result in valid_pairs:
    print(f"w = {w}, 结果 = {result}")
```
