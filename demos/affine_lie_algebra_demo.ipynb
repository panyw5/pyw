# Affine Lie Algebra Computations Demo

This notebook demonstrates the affine Lie algebra utilities following the notation from:

**Di Francesco, Mathieu, Senechal - "Conformal Field Theory", Chapter 14**

## Key Concepts

- **Affine weights**: $\hat{\lambda} = (\lambda; k_\lambda; n_\lambda)$
  - $\lambda$: finite part
  - $k_\lambda$: level (eigenvalue of $\hat{k}$)
  - $n_\lambda$: $L_0$ eigenvalue

- **Affine roots**: $\hat{\alpha} = (\alpha; 0; n) = \alpha + n\delta$

- **Imaginary root**: $\delta = (0; 0; 1)$ with $(\delta, \delta) = 0$

- **Extra simple root**: $\alpha_0 = -\theta + \delta$

## 1. Marks and Comarks

### Marks ($a_i$)
Coefficients in the expansion of the highest root $\theta$:
$$\theta = \sum_{i=0}^r a_i \alpha_i$$

### Comarks ($a_i^\vee$)
Coefficients in the expansion of the highest coroot $\theta^\vee$:
$$\theta^\vee = \sum_{i=0}^r a_i^\vee \alpha_i^\vee$$

```python
# A2^(1) affine Lie algebra - simply laced
ala_a2 = AffineLieAlgebra(['A', 2, 1])

print("=== A2^(1) Affine Lie Algebra ===")
print(f"Marks (a_i):           {ala_a2.get_marks()}")
print(f"Comarks (a_i^vee):     {ala_a2.get_comarks()}")
print(f"Dual Coxeter number:   {ala_a2.dual_coxeter_number()}")
print()

# G2^(1) affine Lie algebra - non-simply laced
ala_g2 = AffineLieAlgebra(['G', 2, 1])

print("=== G2^(1) Affine Lie Algebra ===")
print(f"Marks (a_i):           {ala_g2.get_marks()}")
print(f"Comarks (a_i^vee):     {ala_g2.get_comarks()}")
print(f"Dual Coxeter number:   {ala_g2.dual_coxeter_number()}")
```

## 2. Scalar Products

### Finite Scalar Product $(\lambda, \mu)$
Using the ambient space inner product induced by the Killing form.

### Affine Scalar Product $(\hat{\lambda}, \hat{\mu})$
From Di Francesco Eq. (14.23):
$$(\hat{\lambda}, \hat{\mu}) = (\lambda, \mu) + k_\lambda n_\mu + k_\mu n_\lambda$$

```python
# Get fundamental weights
Lambda = ala_a2.fundamental_weights()

print("=== Scalar Products for A2 ===")
print()

# Finite scalar products
print("Finite scalar products:")
print(f"(L1, L1) = {ala_a2.scalar_product(Lambda[1], Lambda[1])}")
print(f"(L1, L2) = {ala_a2.scalar_product(Lambda[1], Lambda[2])}")
print(f"(L2, L2) = {ala_a2.scalar_product(Lambda[2], Lambda[2])}")
print()

# Affine scalar products
result = ala_a2.affine_scalar_product(Lambda[1], 1, 0, Lambda[2], 1, 0)
print(f"Affine scalar product:")
print(f"(hat_L, hat_M) where hat_L=(L1;1;0), hat_M=(L2;1;0) = {result}")
```

## 3. Weyl Reflections

### Finite Weyl Reflection
$$s_\alpha(\lambda) = \lambda - (\lambda, \alpha^\vee)\alpha$$

### Affine Weyl Reflection (Di Francesco Eq. 14.64)
$$s_{\hat{\alpha}}\hat{\lambda} = (\lambda - [(\lambda,\alpha) + km]\alpha^\vee; k; n - [(\lambda,\alpha) + km]\frac{2m}{|\alpha|^2})$$

```python
# Finite Weyl reflection
alpha = ala_a2.simple_roots()

print("=== Weyl Reflections for A2 ===")
print()

# s1(L1) should give -L1 + L2
result = ala_a2.weyl_reflection(alpha[1], Lambda[1])
print(f"s1(L1) = {result}")
print(f"Expected: -L1 + L2")
print()

# Verify the result
expected = -Lambda[1] + Lambda[2]
print(f"Match: {result == expected}")
print()

# Another example: s2(L2)
result2 = ala_a2.weyl_reflection(alpha[2], Lambda[2])
print(f"s2(L2) = {result2}")
print(f"Expected: -L2 + L1")
```

## 4. Translation Operators

From Di Francesco Eq. (14.69):
$$t_{\alpha^\vee}(\lambda; k; n) = (\lambda + k\alpha^\vee; k; n + \frac{|\lambda|^2 - |\lambda + k\alpha^\vee|^2}{2k})$$

```python
from pyw.core.affine_weight import AffineWeight

alpha_vee = ala_a2.simple_coroots()
zero = ala_a2.weight_lattice().zero()

print("=== Translation Operators for A2^(1) ===")
print()

# Create AffineWeight (0; 1; 0)
w = AffineWeight(ala_a2, zero, level=1, grade=0)

# Translate (0; 1; 0) by alpha1_vee
result = ala_a2.translation(alpha_vee[1], w)

print(f"t_alpha1_vee(0; 1; 0):")
print(f"  New finite part: {result.finite_part}")
print(f"  Level k:         {result.level}")
print(f"  New n:           {result.grade}")
```

## 5. Special Elements

| Element | Description | Value |
|---------|-------------|-------|
| $\delta$ | Imaginary root | $(0; 0; 1)$ with $(\delta, \delta) = 0$ |
| $\theta$ | Highest root of finite algebra | $\theta = \sum a_i \alpha_i$ |
| $\alpha_0$ | Extra simple root | $\alpha_0 = -\theta + \delta$ |
| $\hat{\rho}$ | Affine Weyl vector | $[1, 1, ..., 1]$ |

```python
print("=== Special Elements for A2^(1) ===")
print()

# Imaginary root delta
delta = ala_a2.delta()
print(f"delta = {delta}")
print()

# Highest root theta
theta = ala_a2.theta()
print(f"theta (highest root) = {theta}")
print(f"theta = alpha1 + alpha2: {theta == alpha[1] + alpha[2]}")
print()

# Extra simple root alpha0
alpha_0 = ala_a2.alpha_0()
print(f"alpha0 = {alpha_0}")
print(f"alpha0 = -theta: {alpha_0 == -theta}")
print()

# Affine Weyl vector rho_hat
rho_hat = ala_a2.rho_hat()
print(f"rho_hat = {rho_hat}")
print(f"rho_hat = omega0 + omega1 + omega2: {rho_hat == Lambda[0] + Lambda[1] + Lambda[2]}")
```

## 6. Level and Dynkin Labels

### Level from Dynkin Labels (Di Francesco Eq. 14.54)
$$k = \sum_{i=0}^r a_i^\vee \lambda_i$$

### Zeroth Dynkin Label (Di Francesco Eq. 14.57)
$$\lambda_0 = k - (\lambda, \theta) = k - \sum_{i>0} a_i \lambda_i$$

```python
print("=== Level and Dynkin Labels for A2^(1) ===")
print()

# Level from Dynkin labels
k1 = ala_a2.level_from_dynkin({0: 1, 1: 0, 2: 0})
print(f"Level of [1, 0, 0]: {k1}")

k2 = ala_a2.level_from_dynkin({0: 2, 1: 1, 2: 0})
print(f"Level of [2, 1, 0]: {k2}")
print()

# Zeroth Dynkin label from level
lambda0 = ala_a2.lambda0_from_level(2, {1: 1, 2: 0})
print(f"lambda0 for k=2, finite labels (1: 1, 2: 0): {lambda0}")
print()

# Dominance check
print("Dominance check:")
print(f"[1, 0, 0] is dominant: {ala_a2.is_dominant({0: 1, 1: 0, 2: 0})}")
print(f"[1, 1, 0] is dominant: {ala_a2.is_dominant({0: 1, 1: 1, 2: 0})}")
print(f"[-1, 1, 1] is dominant: {ala_a2.is_dominant({0: -1, 1: 1, 2: 1})}")
```
